```python
bb_data["Column3"] = bb_data["Column3"].replace('[\$,]', '', regex=True).astype(int);
```

```python
plt.plot(bb_data.Column2, bb_data.Column3)
plt.title("BlackBerry Revenue 2009-2022")
plt.xlabel("Year")
plt.ylabel("Millions of USD ($)")
plt.show()
```

```python
nk_data = pd.read_csv("nokiadata.csv")
nk_data["Column3"] = nk_data["Column3"].replace('[\$,]', '', regex=True).astype(int);
```

```python
plt.plot(bb_data.Column2, bb_data.Column3)
plt.plot(nk_data.Column2, nk_data.Column3)
plt.title("Revenue - BlackBerry(BB) vs Nokia(NOK)")
plt.xlabel("Year")
plt.ylabel("Millions of USD ($)")
plt.legend(["Blackberry", "Nokia"])
plt.show()
```
